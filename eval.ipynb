# Evaluation Metrics

Evaluation metrics measure how well a machine learning model performs on unseen data.

The choice of metric depends on:

- Classification vs Regression
- Balanced vs Imbalanced Data
- Business Objectives

---

# Classification Metrics

Used when the target variable is categorical.

Examples:

```text
Spam vs Not Spam
Fraud vs Non-Fraud
Disease vs Healthy
```

## Confusion Matrix

Most classification metrics are derived from the confusion matrix.

| Actual / Predicted | Positive | Negative |
|------------|------------|------------|
| Positive | TP | FN |
| Negative | FP | TN |

Where:

- TP = True Positive
- TN = True Negative
- FP = False Positive
- FN = False Negative

---

## Accuracy

Measures overall correctness.

$$
Accuracy
=
\frac{TP+TN}
{TP+TN+FP+FN}
$$

### Example

```text
100 predictions
90 correct
10 incorrect
```

$$
Accuracy = 90\%
$$

### Pros

- Easy to understand

### Cons

Poor metric for imbalanced datasets.

Example:

```text
99% Non-Fraud
1% Fraud
```

Predicting everything as non-fraud gives:

```text
Accuracy = 99%
```

while detecting no fraud.

---

## Precision

Measures prediction quality.

$$
Precision
=
\frac{TP}
{TP+FP}
$$

Question:

> Of all predicted positives, how many were actually positive?

### Example

```text
Predicted Fraud = 100
Actual Fraud = 80
```

$$
Precision = 80\%
$$

### Important When

False Positives are costly.

Examples:

- Spam filtering
- Fraud alerts

---

## Recall

Measures detection ability.

$$
Recall
=
\frac{TP}
{TP+FN}
$$

Question:

> Of all actual positives, how many did we find?

### Example

```text
Actual Fraud = 100
Detected Fraud = 90
```

$$
Recall = 90\%
$$

### Important When

False Negatives are costly.

Examples:

- Cancer Detection
- Fraud Detection
- Security Systems

---

## F1 Score

Balances Precision and Recall.

$$
F1
=
2
\frac{Precision \times Recall}
{Precision + Recall}
$$

### Why Use F1?

Accuracy can be misleading for imbalanced datasets.

F1 considers both:

```text
Precision
Recall
```

### Use When

- Classes are imbalanced
- Precision and Recall are equally important

---

## ROC-AUC

ROC = Receiver Operating Characteristic

AUC = Area Under Curve

ROC plots:

```text
True Positive Rate
vs
False Positive Rate
```

AUC measures ranking quality.

### Interpretation

| AUC | Meaning |
|------|----------|
| 1.0 | Perfect |
| 0.9+ | Excellent |
| 0.8+ | Good |
| 0.5 | Random Guessing |

### Pros

- Threshold independent
- Measures ranking performance

### Cons

Can be misleading for highly imbalanced datasets.

---

## PR-AUC

PR = Precision-Recall Curve

Plots:

```text
Precision
vs
Recall
```

### Why Use PR-AUC?

For imbalanced datasets:

```text
Fraud Detection
Medical Diagnosis
Rare Event Prediction
```

PR-AUC is often more informative than ROC-AUC.

### Rule of Thumb

```text
Balanced Data
→ ROC-AUC

Imbalanced Data
→ PR-AUC
```

---

# Regression Metrics

Used when the target variable is continuous.

Examples:

```text
House Price
Salary
Sales Forecast
Temperature
```

---

## MAE (Mean Absolute Error)

Measures average absolute prediction error.

$$
MAE
=
\frac1n
\sum |y-\hat y|
$$

### Example

```text
Actual = 100
Predicted = 90

Error = 10
```

### Advantages

- Easy to interpret
- Same units as target

### Disadvantages

- Does not strongly penalize large errors

---

## MSE (Mean Squared Error)

Measures average squared error.

$$
MSE
=
\frac1n
\sum(y-\hat y)^2
$$

### Advantages

- Penalizes large errors heavily
- Smooth optimization surface

### Disadvantages

- Sensitive to outliers

---

## RMSE (Root Mean Squared Error)

Square root of MSE.

$$
RMSE
=
\sqrt{MSE}
$$

### Advantages

- Same unit as target variable
- Penalizes large errors

### Interpretation

```text
RMSE = 5
```

means predictions are off by about 5 units.

---

## R² (Coefficient of Determination)

Measures variance explained by the model.

$$
R^2
=
1-
\frac{SS_{res}}
{SS_{tot}}
$$

where:

- \(SS_{res}\) = Residual Sum of Squares
- \(SS_{tot}\) = Total Sum of Squares

### Interpretation

| R² | Meaning |
|------|----------|
| 1.0 | Perfect Fit |
| 0.8 | Explains 80% of variance |
| 0 | Same as predicting mean |
| <0 | Worse than mean prediction |

### Advantages

- Easy interpretation
- Widely reported

---

# Ranking Metrics

Used when ranking items is more important than predicting exact labels.

Applications:

- Search Engines
- Recommendation Systems
- Ads Ranking
- Feed Ranking

---

## MAP (Mean Average Precision)

Measures ranking quality across multiple queries.

### Idea

Reward:

```text
Relevant items appearing early
```

Average Precision is computed for each query.

MAP is the average across all queries.

### Use Cases

- Search Engines
- Information Retrieval

---

## MRR (Mean Reciprocal Rank)

Focuses on the first correct result.

$$
MRR
=
\frac1N
\sum
\frac1{Rank}
$$

### Example

If first relevant result appears at:

```text
Position 1 → Score = 1
Position 2 → Score = 0.5
Position 5 → Score = 0.2
```

### Use Cases

- Question Answering
- Search
- Chatbots

---

## NDCG (Normalized Discounted Cumulative Gain)

One of the most important ranking metrics.

Rewards:

```text
Relevant Items
+
Higher Positions
```

Idea:

A relevant item at position 1 is more valuable than at position 20.

### Advantages

- Supports multiple relevance levels
- Measures ranking quality well

### Use Cases

- Search Ranking
- Recommendation Systems
- TikTok Feed Ranking
- YouTube Ranking

---

## Hit Rate

Measures whether at least one relevant item appears in the top K results.

$$
HitRate@K
=
\frac{\text{Hits}}
{\text{Total Users}}
$$

### Example

Recommend:

```text
Top 10 Movies
```

If user likes at least one:

```text
Hit = 1
```

Otherwise:

```text
Hit = 0
```

### Use Cases

- Recommender Systems
- Candidate Retrieval
- Two-Tower Models

---

# Metric Selection Guide

| Problem | Recommended Metrics |
|-----------|-----------|
| Balanced Classification | Accuracy, ROC-AUC |
| Imbalanced Classification | Precision, Recall, F1, PR-AUC |
| Medical Diagnosis | Recall, PR-AUC |
| Spam Detection | Precision |
| House Price Prediction | MAE, RMSE, R² |
| Forecasting | MAE, RMSE |
| Search Ranking | MAP, MRR, NDCG |
| Recommendation Systems | NDCG, Hit Rate |

---

# Interview Summary

### Classification

- Accuracy → Overall correctness
- Precision → Quality of positive predictions
- Recall → Ability to find positives
- F1 → Balance of Precision and Recall
- ROC-AUC → Ranking quality
- PR-AUC → Best for imbalanced data

### Regression

- MAE → Average absolute error
- MSE → Squared error
- RMSE → Root squared error
- R² → Variance explained

### Ranking

- MAP → Ranking quality across queries
- MRR → Position of first relevant result
- NDCG → Ranking quality with position importance
- Hit Rate → Whether relevant items appear in Top-K
