# 7. Evaluation Metrics

Evaluation metrics measure how well a machine learning model performs on unseen data.

The choice of metric depends on:

- Classification vs Regression
- Balanced vs Imbalanced Data
- Business Objectives

---

# Classification Metrics

Used when the target variable is categorical.

Examples:

```text
Spam vs Not Spam
Fraud vs Non-Fraud
Disease vs Healthy
```

---

## Confusion Matrix

Most classification metrics are derived from the confusion matrix.

| Actual / Predicted | Positive | Negative |
|------------|------------|------------|
| Positive | TP | FN |
| Negative | FP | TN |

Where:

- TP = True Positive
- TN = True Negative
- FP = False Positive
- FN = False Negative

---

## Accuracy

Measures overall correctness.

$$
Accuracy
=
\frac{TP+TN}
{TP+TN+FP+FN}
$$

### Example

```text
100 predictions
90 correct
10 incorrect
```

$$
Accuracy = 90\%
$$

### Pros

- Easy to understand

### Cons

Poor metric for imbalanced datasets.

Example:

```text
99% Non-Fraud
1% Fraud
```

Predicting everything as non-fraud gives:

```text
Accuracy = 99%
```

while detecting no fraud.

---

## Precision

Measures prediction quality.

$$
Precision
=
\frac{TP}
{TP+FP}
$$

Question:

> Of all predicted positives, how many were actually positive?

### Example

```text
Predicted Fraud = 100
Actual Fraud = 80
```

$$
Precision = 80\%
$$

### Important When

False Positives are costly.

Examples:

- Spam Filtering
- Fraud Alerts
- Loan Approval

---

## Recall

Measures detection ability.

$$
Recall
=
\frac{TP}
{TP+FN}
$$

Question:

> Of all actual positives, how many did we find?

### Example

```text
Actual Fraud = 100
Detected Fraud = 90
```

$$
Recall = 90\%
$$

### Important When

False Negatives are costly.

Examples:

- Cancer Detection
- Fraud Detection
- Security Systems

---

## F1 Score

Balances Precision and Recall.

$$
F1
=
2
\frac{Precision \times Recall}
{Precision + Recall}
$$

### Why Use F1?

Accuracy can be misleading for imbalanced datasets.

F1 considers both:

```text
Precision
Recall
```

### Use When

- Classes are imbalanced
- Precision and Recall are equally important

---

# ROC-AUC (Receiver Operating Characteristic - Area Under Curve)

## Why Do We Need ROC-AUC?

Most classifiers output probabilities.

Example:

| Customer | Fraud Probability |
|-----------|-----------|
| A | 0.95 |
| B | 0.80 |
| C | 0.60 |
| D | 0.30 |
| E | 0.10 |

To make predictions we choose a threshold.

Example:

```text
Threshold = 0.5
```

Then:

```text
Probability > 0.5 → Fraud
Probability ≤ 0.5 → Not Fraud
```

Problem:

Different thresholds produce different results.

ROC-AUC evaluates the model across all possible thresholds.

---

## ROC Curve

ROC stands for:

```text
Receiver Operating Characteristic
```

The ROC Curve plots:

```text
True Positive Rate (TPR)
vs
False Positive Rate (FPR)
```

---

### True Positive Rate (TPR)

Also called Recall.

$$
TPR
=
\frac{TP}
{TP+FN}
$$

Question:

> Of all actual positives, how many did we correctly identify?

---

### False Positive Rate (FPR)

$$
FPR
=
\frac{FP}
{FP+TN}
$$

Question:

> Of all actual negatives, how many were incorrectly predicted as positive?

---

## ROC Curve Example

Suppose:

```text
Threshold = 0.9
```

Very strict.

```text
TPR = Low
FPR = Very Low
```

---

Lower threshold:

```text
Threshold = 0.5
```

```text
TPR Increases
FPR Increases
```

---

Very low threshold:

```text
Threshold = 0.1
```

```text
TPR ≈ 1
FPR ≈ 1
```

---

ROC Curve:

```text
TPR
1.0 |               *
    |            *
    |         *
    |      *
    |   *
0.0 +-------------------
     0.0            1.0
           FPR
```

Each point corresponds to a different threshold.

---

## What is AUC?

AUC = Area Under the ROC Curve

It measures ranking quality.

### Interpretation

| AUC | Meaning |
|------|----------|
| 1.0 | Perfect |
| 0.9+ | Excellent |
| 0.8+ | Good |
| 0.7+ | Fair |
| 0.5 | Random Guessing |
| < 0.5 | Worse than Random |

---

## Intuition Behind AUC

AUC can be interpreted as:

> The probability that a randomly chosen positive example receives a higher score than a randomly chosen negative example.

Example:

| Example | Score |
|----------|----------|
| Fraud | 0.90 |
| Non-Fraud | 0.20 |

Positive ranked above negative.

Good.

---

Another pair:

| Example | Score |
|----------|----------|
| Fraud | 0.40 |
| Non-Fraud | 0.80 |

Negative ranked above positive.

Bad.

---

AUC measures how often the model ranks positives above negatives.

---

## Perfect ROC Curve

Perfect classifier:

```text
TPR = 1
FPR = 0
```

AUC:

$$
AUC = 1
$$

---

## Random Classifier

Random guessing:

```text
TPR = FPR
```

AUC:

$$
AUC = 0.5
$$

---

## Why ROC-AUC is Better than Accuracy

Example:

```text
1000 Customers
10 Fraud
990 Non-Fraud
```

Predict:

```text
Everything = Non-Fraud
```

Accuracy:

$$
99\%
$$

Looks great.

Reality:

```text
Detected Fraud = 0
```

ROC-AUC would expose the poor ranking performance.

---

## Advantages

- Threshold independent
- Measures ranking quality
- Widely used for binary classification

### Limitations

Can be overly optimistic for highly imbalanced datasets.

---

## Answer

ROC-AUC measures how well a model separates positive and negative classes across all possible thresholds. It can be interpreted as the probability that a randomly chosen positive sample receives a higher score than a randomly chosen negative sample.

---

## PR-AUC

PR = Precision-Recall Curve

Plots:

```text
Precision
vs
Recall
```

### Why Use PR-AUC?

For imbalanced datasets:

```text
Fraud Detection
Medical Diagnosis
Rare Event Prediction
```

PR-AUC focuses on the positive class and is often more informative than ROC-AUC.

### Rule of Thumb

```text
Balanced Data
→ ROC-AUC

Imbalanced Data
→ PR-AUC
```

---

# Regression Metrics

Used when the target variable is continuous.

Examples:

```text
House Price
Salary
Sales Forecast
Temperature
```

---

## MAE (Mean Absolute Error)

Measures average absolute prediction error.

$$
MAE
=
\frac1n
\sum |y-\hat y|
$$

### Advantages

- Easy to interpret
- Same units as target

### Disadvantages

- Does not strongly penalize large errors

---

## MSE (Mean Squared Error)

Measures average squared error.

$$
MSE
=
\frac1n
\sum(y-\hat y)^2
$$

### Advantages

- Penalizes large errors heavily
- Smooth optimization objective

### Disadvantages

- Sensitive to outliers

---

## RMSE (Root Mean Squared Error)

Square root of MSE.

$$
RMSE
=
\sqrt{MSE}
$$

### Advantages

- Same unit as target variable
- Penalizes large errors

### Interpretation

```text
RMSE = 5
```

Predictions are off by about 5 units on average.

---

## R² (Coefficient of Determination)

Measures variance explained by the model.

$$
R^2
=
1-
\frac{SS_{res}}
{SS_{tot}}
$$

where:

- \(SS_{res}\) = Residual Sum of Squares
- \(SS_{tot}\) = Total Sum of Squares

### Interpretation

| R² | Meaning |
|------|----------|
| 1.0 | Perfect Fit |
| 0.8 | Explains 80% of variance |
| 0 | Same as predicting mean |
| <0 | Worse than predicting mean |

---

# Ranking Metrics

Used when ranking items is more important than predicting exact labels.

Applications:

- Search Engines
- Recommendation Systems
- Ads Ranking
- Social Media Feed Ranking

---

## MAP (Mean Average Precision)

Measures ranking quality across multiple queries.

Idea:

```text
Relevant items appearing early
```

Higher MAP means relevant results are ranked near the top.

Applications:

- Search Engines
- Information Retrieval

---

## MRR (Mean Reciprocal Rank)

Focuses on the position of the first relevant result.

$$
MRR
=
\frac1N
\sum
\frac1{Rank}
$$

Example:

| First Relevant Result | Score |
|------|------|
| Rank 1 | 1.0 |
| Rank 2 | 0.5 |
| Rank 5 | 0.2 |

Applications:

- Search
- Question Answering
- Chatbots

---

## NDCG (Normalized Discounted Cumulative Gain)

Rewards:

```text
Relevant Items
+
Higher Positions
```

A relevant item at position 1 is more valuable than position 20.

Applications:

- Search Ranking
- Recommendation Systems
- TikTok Feed Ranking
- YouTube Ranking

Most commonly used ranking metric in recommender systems.

---

## Hit Rate

Measures whether at least one relevant item appears in the top K recommendations.

$$
HitRate@K
=
\frac{\text{Hits}}
{\text{Total Users}}
$$

Example:

```text
Recommend Top 10 Movies
```

If user likes at least one:

```text
Hit = 1
```

Otherwise:

```text
Hit = 0
```

Applications:

- Recommender Systems
- Candidate Retrieval
- Two-Tower Retrieval Models

---

# Metric Selection Guide

| Problem | Recommended Metrics |
|-----------|-----------|
| Balanced Classification | Accuracy, ROC-AUC |
| Imbalanced Classification | Precision, Recall, F1, PR-AUC |
| Medical Diagnosis | Recall, PR-AUC |
| Spam Detection | Precision |
| House Price Prediction | MAE, RMSE, R² |
| Forecasting | MAE, RMSE |
| Search Ranking | MAP, MRR, NDCG |
| Recommendation Systems | NDCG, Hit Rate |

---


### Classification

- Accuracy → Overall correctness
- Precision → How accurate positive predictions are
- Recall → How many positives were found
- F1 → Balance between Precision and Recall
- ROC-AUC → Ranking quality across all thresholds
- PR-AUC → Best metric for imbalanced datasets

### Regression

- MAE → Average absolute error
- MSE → Squared error
- RMSE → Root squared error
- R² → Variance explained

### Ranking

- MAP → Ranking quality across queries
- MRR → Position of first relevant result
- NDCG → Ranking quality with position importance
- Hit Rate → Whether relevant items appear in Top-K
