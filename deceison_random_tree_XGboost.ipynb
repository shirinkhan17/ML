
# Tree-Based Models

Tree models recursively split data into smaller groups.

Advantages:

- Handle nonlinear relationships
- Minimal preprocessing
- Feature importance available

---



# Decision Tree vs Random Forest vs XGBoost

## Quick Comparison

| Aspect | Decision Tree | Random Forest | XGBoost |
|----------|----------|----------|----------|
| Model Type | Single Tree | Bagging Ensemble | Boosting Ensemble |
| Number of Trees | 1 | Many Trees | Many Trees |
| Training Style | Single Tree | Parallel Training | Sequential Training |
| Main Goal | Learn Decision Rules | Reduce Variance | Reduce Bias + Optimize Loss |
| Cost Function | Gini / Entropy / MSE | Same as Decision Tree | Loss + Regularization |
| Regularization | Pruning | Pruning + Feature Randomness | L1, L2, Gamma, Subsampling |
| Bias | Low | Medium-Low | Very Low |
| Variance | High | Low | Medium |
| Overfitting Risk | High | Low | Medium-High |
| Interpretability | High | Medium | Low |
| Training Speed | Fast | Medium | Slower |
| Hyperparameter Tuning | Easy | Moderate | Extensive |
| Typical Accuracy | Lowest | High | Usually Highest |
| Handles Missing Values | Limited | Limited | Native Support |

# 1. Decision Tree

## Structure

```text
             Root
            /    \
        Node      Node
        / \       / \
     Leaf Leaf  Leaf Leaf
```

A prediction follows a single path from root to leaf.

---

## Cost Function

### Classification

**Gini Impurity**

$$
Gini = 1 - \sum p_i^2
$$

**Entropy**

$$
Entropy = -\sum p_i \log p_i
$$

Choose the split that maximizes impurity reduction.

### Regression

**Mean Squared Error (MSE)**

$$
MSE = \frac{1}{N}\sum (y_i-\bar y)^2
$$

Choose the split that minimizes variance.

---

## Regularization

| Parameter | Purpose |
|------------|----------|
| max_depth | Limit depth |
| min_samples_split | Minimum samples before splitting |
| min_samples_leaf | Minimum samples in leaf |
| max_leaf_nodes | Limit number of leaves |
| ccp_alpha | Cost-complexity pruning |

---

## Bias and Variance

### Why Low Bias?

Trees can learn complex nonlinear patterns.

### Why High Variance?

Small changes in data can create completely different trees.

Example:

Dataset A:

```text
Age < 30?
```

Dataset B:

```text
Income < 50K?
```

Different root split → different tree → different prediction.

---

# 2. Random Forest

## Structure

```text
            Data
              |
      ----------------
      |  |  |  |  |  |
     T1 T2 T3 T4 T5 T6
      |  |  |  |  |  |
      ----------------
              |
        Average/Vote
```

Many independent trees trained in parallel.

---

## Cost Function

Each tree independently optimizes:

### Classification

- Gini
- Entropy

### Regression

- MSE

No global forest objective exists.

---

## Prediction

### Classification

Majority Vote

$$
\hat y = mode(T_1,T_2,\ldots,T_n)
$$

### Regression

Average Prediction

$$
\hat y = \frac{1}{n}\sum T_i
$$

---

## Regularization

| Parameter | Purpose |
|------------|----------|
| max_depth | Limit tree complexity |
| min_samples_leaf | Prevent tiny leaves |
| max_features | Reduce correlation between trees |
| max_samples | Sample fewer rows |

---

## Why Random Forest Has Lower Variance

Suppose:

| Tree | Prediction |
|--------|------------|
| T1 | 90 |
| T2 | 110 |
| T3 | 95 |
| T4 | 105 |
| T5 | 100 |

Average:

$$
100
$$

Random errors cancel out.

Variance decreases approximately as:

$$
Var\left(\frac1n \sum T_i\right)
=
\frac{\sigma^2}{n}
$$

(assuming low correlation among trees)

---

# 3. XGBoost

## Structure

```text
Tree1
  ↓
Residual Error
  ↓
Tree2
  ↓
Residual Error
  ↓
Tree3
  ↓
Final Prediction
```

Each new tree learns the mistakes of previous trees.

---

## Prediction

$$
\hat y
=
T_1 + T_2 + T_3 + \cdots + T_n
$$

---

## Objective Function

$$
Objective
=
Loss
+
Regularization
$$

### Classification

$$
LogLoss + Regularization
$$

### Regression

$$
MSE + Regularization
$$

---

## Regularization

### Tree Complexity

| Parameter | Purpose |
|------------|----------|
| max_depth | Limit depth |
| gamma | Minimum gain for split |
| min_child_weight | Prevent tiny leaves |

### Sampling

| Parameter | Purpose |
|------------|----------|
| subsample | Sample rows |
| colsample_bytree | Sample columns |

### L1 and L2

$$
\Omega(f)
=
\gamma T
+
\frac12 \lambda \sum w_j^2
+
\alpha \sum |w_j|
$$

Where:

- reg_alpha = L1
- reg_lambda = L2

### Shrinkage

$$
learning\_rate = \eta
$$

### Early Stopping

Stop training when validation performance stops improving.

---

## Why XGBoost Has Low Bias

Example:

Tree 1:

```text
Actual = 100
Prediction = 80
Error = 20
```

Tree 2 learns error:

```text
80 + 15 = 95
```

Tree 3:

```text
95 + 4 = 99
```

Tree 4:

```text
99 + 1 = 100
```

Each tree corrects previous mistakes.

---

## Why XGBoost Can Overfit

Too many trees may start fitting noise:

```text
Tree1 → Signal
Tree2 → Signal
Tree20 → Signal
Tree100 → Noise
```

Training Error ↓

Validation Error ↑

This is why regularization and early stopping are important.

# Summary

### Decision Tree

- Single tree
- Low bias
- High variance
- Easy to interpret

### Random Forest

- Bagging
- Parallel trees
- Reduces variance through averaging
- Strong baseline model

### XGBoost

- Boosting
- Sequential trees
- Reduces bias by correcting errors
- Global objective = Loss + Regularization
- Usually highest accuracy on tabular data

### One-Line Answer

> Decision Trees suffer from high variance because small changes in training data can produce very different trees. Random Forest reduces variance by averaging many decorrelated trees trained in parallel. XGBoost reduces bias by sequentially fitting residual errors and optimizing a regularized objective function, often achieving the highest predictive accuracy on tabular datasets.


## Decision Tree

Structure:

```text
Root
 ├─ Node
 │   ├─ Leaf
 │   └─ Leaf
 └─ Node
     ├─ Leaf
     └─ Leaf
```

Split Criteria:

- Gini Impurity
- Entropy
- MSE (Regression)

Pros:

- Easy to interpret
- Handles nonlinear patterns

Cons:

- High variance
- Overfits easily

---

## Random Forest

Ensemble of many decision trees.

Key Ideas:

- Bootstrap Sampling
- Feature Randomness
- Averaging Predictions

Pros:

- Lower variance than Decision Trees
- Strong baseline model

Cons:

- Less interpretable
- Larger memory usage

---

## Gradient Boosting

Builds trees sequentially.

Each new tree learns the errors of previous trees.

```text
Tree 1
   ↓
Residual Error
   ↓
Tree 2
   ↓
Residual Error
   ↓
Tree 3
```

Pros:

- High predictive power

Cons:

- Can overfit without tuning

---

## XGBoost

Optimized implementation of Gradient Boosting.

Features:

- Regularization
- Learning Rate
- Tree Pruning
- Parallel Processing
- Early Stopping

Pros:

- Often best performance on tabular data

Cons:

- More hyperparameters

---

## LightGBM

Gradient Boosting framework developed by Microsoft.

Features:

- Histogram-based training
- Leaf-wise growth

Pros:

- Very fast
- Scales to large datasets

Cons:

- More prone to overfitting on small datasets

---

## CatBoost

Gradient Boosting framework developed by Yandex.

Features:

- Native handling of categorical features
- Ordered boosting

Pros:

- Excellent for categorical data
- Minimal preprocessing

Cons:

- Slower than LightGBM
