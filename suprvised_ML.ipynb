# 2. Supervised Learning

## What is Supervised Learning?

Supervised Learning is a machine learning approach where a model learns from labeled data.

Training data contains:

$$
(X, Y)
$$

where:

- \(X\) = input features
- \(Y\) = target labels

Goal:

Learn a function:

$$
Y = f(X)
$$

that can predict outcomes for unseen data.

---

## Types of Supervised Learning

### Classification

Predict discrete categories.

Examples:

- Spam vs Not Spam
- Fraud vs Non-Fraud
- Disease vs Healthy

Common Algorithms:

- Logistic Regression
- Decision Tree
- Random Forest
- XGBoost
- SVM
- Neural Networks

---

### Regression

Predict continuous values.

Examples:

- House Price
- Sales Forecasting
- Temperature Prediction

Common Algorithms:

- Linear Regression
- Random Forest Regressor
- XGBoost Regressor
- Neural Networks

---

# Linear Models

Linear models assume a linear relationship between features and target.

Advantages:

- Simple
- Fast
- Interpretable

Limitations:

- Cannot easily model complex nonlinear relationships

---

## Linear Regression

Used for regression problems.

Model:

$$
y = w_1x_1 + w_2x_2 + ... + w_nx_n + b
$$

Cost Function:

$$
MSE = \frac{1}{n}\sum (y-\hat y)^2
$$

Pros:

- Simple
- Fast
- Highly interpretable

Cons:

- Assumes linear relationships
- Sensitive to outliers

---

## Logistic Regression

Used for classification problems.

Prediction:

$$
p = \frac{1}{1+e^{-z}}
$$

where:

$$
z = wx+b
$$

Output:

```text
0 → Negative Class
1 → Positive Class
```

Cost Function:

Binary Cross Entropy (Log Loss)

Pros:

- Fast
- Interpretable
- Strong baseline classifier

Cons:

- Limited for highly nonlinear data

---

## Naive Bayes

Probabilistic classifier based on Bayes' theorem.

Assumption:

Features are conditionally independent.


Variants:

- Gaussian Naive Bayes
- Multinomial Naive Bayes
- Bernoulli Naive Bayes

Pros:

- Extremely fast
- Works well on text data

Cons:

- Independence assumption is often unrealistic

---



## Support Vector Machine (SVM)

Goal:

Find the decision boundary with the maximum margin.

Concepts:

- Maximum Margin
- Soft Margin
- Kernel Trick

Common Kernels:

- Linear
- Polynomial
- RBF

Pros:

- Effective in high-dimensional spaces

Cons:

- Slow on large datasets

---

## K-Nearest Neighbors (KNN)

Prediction based on nearby observations.

Algorithm:

1. Choose K neighbors
2. Compute distances
3. Aggregate neighbor labels

Distance Metrics:

- Euclidean Distance
- Manhattan Distance

Pros:

- Simple
- No training phase

Cons:

- Slow during prediction
- Suffers from curse of dimensionality

---

# Model Comparison

| Model | Interpretable | Handles Nonlinearity | Fast Training | Typical Accuracy |
|---------|-------------|-------------|-------------|-------------|
| Linear Regression | Yes | No | Very Fast | Medium |
| Logistic Regression | Yes | Limited | Very Fast | Medium |
| Naive Bayes | Moderate | Limited | Very Fast | Medium |
| Decision Tree | Yes | Yes | Fast | Medium |
| Random Forest | Partial | Yes | Medium | High |
| XGBoost | No | Yes | Medium | Very High |
| LightGBM | No | Yes | Fast | Very High |
| CatBoost | No | Yes | Medium | Very High |
| SVM | Partial | Yes | Slow | High |
| KNN | Moderate | Yes | No Training | Medium |

---

## Key Summary

- Linear Regression → regression problems.
- Logistic Regression → classification problems.
- Naive Bayes → probabilistic classifier.
- Decision Tree → interpretable but high variance.
- Random Forest → reduces variance through averaging.
- Gradient Boosting → sequential error correction.
- XGBoost → regularized boosting with high accuracy.
- LightGBM → fast gradient boosting for large data.
- CatBoost → best for categorical features.
- SVM → maximum-margin classifier.
- KNN → instance-based learning using nearest neighbors.
