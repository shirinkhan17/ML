# 1. Machine Learning Fundamentals

## What is Machine Learning?

Machine Learning is a field of Artificial Intelligence where computers learn patterns from data and make predictions or decisions without being explicitly programmed.

Traditional programming:

```text
Input + Rules → Output
```

Machine learning:

```text
Input + Output → Learn Rules / Model
```

The goal is to learn a function:

$$
y = f(x)
$$

where:

- \(x\) = input features
- \(y\) = target/output
- \(f(x)\) = learned model

## Cost Function and Gradient

Machine learning models learn by minimizing a **cost function** (also called a loss function).

### What is a Cost Function?

A cost function measures how far the model's predictions are from the true values.

Goal:

```text
Small Cost → Good Model
Large Cost → Poor Model
```

The learning process tries to find model parameters that minimize the cost.

---

### Example: Linear Regression

Suppose:

$$
\hat y = wx + b
$$

where:

- \(w\) = weight
- \(b\) = bias
- \(\hat y\) = prediction

The most common cost function is Mean Squared Error (MSE):

$$
J(w,b)
=
\frac{1}{n}
\sum_{i=1}^{n}
(y_i-\hat y_i)^2
$$

where:

- \(y_i\) = actual value
- \(\hat y_i\) = predicted value

If predictions are perfect:

$$
J=0
$$

Otherwise:

$$
J>0
$$

---

### What is a Gradient?

A gradient tells us:

> "In which direction should we move the parameters to reduce the cost fastest?"

Mathematically, it is the derivative of the cost function.

For a parameter \(w\):

$$
\frac{\partial J}{\partial w}
$$

Interpretation:

- Positive gradient → move left
- Negative gradient → move right
- Large magnitude → steep slope
- Near zero → close to optimum

---

### Intuition

Imagine standing on a mountain.

```text
        *
      *   *
    *       *
  *           *
--------------------
```

Your goal is to reach the lowest point.

The gradient tells you:

```text
"Walk downhill this way."
```

Gradient Descent repeatedly follows the downhill direction until reaching a minimum.

---

### Gradient Descent Update Rule

Parameters are updated as:


$$
w_{new} = w_{old} - \eta \frac{\partial J}{\partial w}
$$

where:

- $\eta$ = learning rate
- $\frac{\partial J}{\partial w}$ = gradient of the cost function with respect to \(w\)

Similarly:

$$
b_{new} = b_{old} - \eta \frac{\partial J}{\partial b}
$$

---

### Why Learning Rate Matters

#### Too Small

```text
Very Slow Learning
```

Many iterations needed.

#### Too Large

```text
Jump Over Minimum
```

Training becomes unstable.

#### Good Learning Rate

```text
Fast and Stable Convergence
```

---

### Gradient Descent Process

```text
Initialize Parameters
          ↓
Compute Predictions
          ↓
Compute Cost
          ↓
Compute Gradients
          ↓
Update Parameters
          ↓
Repeat Until Convergence
```

---

### Common Cost Functions

#### Regression

Mean Squared Error (MSE)

$$
MSE
=
\frac{1}{n}
\sum (y-\hat y)^2
$$

Mean Absolute Error (MAE)

$$
MAE
=
\frac{1}{n}
\sum |y-\hat y|
$$

---

#### Classification

Binary Cross Entropy (Log Loss)

$$
-\left[
y\log(p)
+
(1-y)\log(1-p)
\right]
$$

Used in:

- Logistic Regression
- Neural Networks

---

### Summary

- Cost function measures prediction error.
- Training aims to minimize the cost function.
- Gradient is the derivative of the cost function.
- Gradient indicates the direction of steepest increase.
- Gradient Descent moves in the opposite direction of the gradient.
- Learning rate controls step size.
- Common losses:
  - MSE for regression
  - Cross Entropy for classification
## Bias vs Variance

### Bias

Bias is error from a model being too simple.

High bias means the model cannot capture the true pattern.

Example:

- Linear model on nonlinear data
- Very shallow decision tree

Symptoms:

```text
Training Error: High
Validation Error: High
```

This is called **underfitting**.

---

### Variance

Variance is error from a model being too sensitive to training data.

High variance means the model learns noise instead of general patterns.

Example:

- Very deep decision tree
- Very complex neural network

Symptoms:

```text
Training Error: Low
Validation Error: High
```

This is called **overfitting**.

---

### Bias-Variance Tradeoff

$$
Total Error = Bias^2 + Variance + Irreducible\ Noise
$$

Goal:

```text
Low Bias + Low Variance
```

---

## Underfitting vs Overfitting

| Concept | Meaning | Training Error | Validation Error |
|---|---|---|---|
| Underfitting | Model too simple | High | High |
| Good Fit | Model generalizes well | Low | Low |
| Overfitting | Model too complex | Very Low | High |

### Underfitting

The model misses important patterns.

Causes:

- Too simple model
- Too much regularization
- Not enough features
- Too few training epochs

### Overfitting

The model memorizes training data.

Causes:

- Too complex model
- Too little data
- Too many features
- Too many training epochs

## Train / Validation / Test Split

Data is usually divided into three parts.

| Dataset | Purpose |
|---|---|
| Training Set | Used to train the model |
| Validation Set | Used to tune hyperparameters |
| Test Set | Used for final evaluation |

Common split:

```text
Training: 70%
Validation: 15%
Test: 15%
```

Important rule:

> Do not tune the model using the test set.

The test set should be used only once for final evaluation.

## Cross Validation

Cross validation is a technique used to evaluate how well a model is likely to perform on unseen data.

Instead of splitting the data only once into training and validation sets, we train and evaluate the model multiple times using different subsets of the data.

### Why Do We Need It?

Suppose we have 1000 samples.

If we create one train-validation split:

```text
Train = 800
Validation = 200
```

our validation score depends on which 200 samples happened to be selected.

Example:

```text
Split 1 → Accuracy = 92%
Split 2 → Accuracy = 86%
Split 3 → Accuracy = 89%
```

The score can vary because different validation sets contain different data.

Cross validation reduces this dependency on a single split.

---

### K-Fold Cross Validation

In K-Fold Cross Validation, the dataset is divided into K equal parts (folds).

Example: 5-Fold Cross Validation

```text
Fold 1 | Fold 2 | Fold 3 | Fold 4 | Fold 5
```

Training and validation are repeated 5 times:

```text
Run 1:
Train = Fold 2+3+4+5
Validate = Fold 1

Run 2:
Train = Fold 1+3+4+5
Validate = Fold 2

Run 3:
Train = Fold 1+2+4+5
Validate = Fold 3

Run 4:
Train = Fold 1+2+3+5
Validate = Fold 4

Run 5:
Train = Fold 1+2+3+4
Validate = Fold 5
```

Every sample is used:

- For training K−1 times
- For validation 1 time

---

### Final Score

Suppose the model achieves:

```text
Fold 1 = 91%
Fold 2 = 89%
Fold 3 = 90%
Fold 4 = 92%
Fold 5 = 88%
```

Final score:

$$
CV = \frac{91 + 89 + 90 + 92 + 88}{5}
= 90\%
$$

---

### Advantages

- Uses all data efficiently
- Less sensitive to one lucky or unlucky split
- More reliable estimate of model performance
- Useful when the dataset is small

### Disadvantages

- Slower because the model is trained multiple times
- More computationally expensive

---

### NOTES

Cross validation evaluates a model on multiple train-validation splits instead of a single split. This reduces the variance of the performance estimate and provides a more reliable measure of how the model will generalize to unseen data.

### K-Fold Cross Validation

The data is split into \(k\) folds.

Example with 5 folds:

```text
Fold 1 → Validation, others → Training
Fold 2 → Validation, others → Training
Fold 3 → Validation, others → Training
Fold 4 → Validation, others → Training
Fold 5 → Validation, others → Training
```

Final score:

$$
CV\ Score = \frac{1}{k}\sum_{i=1}^{k} Score_i
$$

Benefits:

- Better use of data
- More stable model evaluation
- Useful for small datasets

## Regularization

Regularization reduces overfitting by penalizing model complexity.

### L1 Regularization

Also called **Lasso**.

$$
Loss + \lambda \sum |w_i|
$$

Effect:

- Pushes some weights to exactly zero
- Performs feature selection
- Creates sparse models

---

### L2 Regularization

Also called **Ridge**.

$$
Loss + \lambda \sum w_i^2
$$

Effect:

- Shrinks weights toward zero
- Keeps most features
- Creates more stable models

---

### Other Regularization Methods

| Method | Common Use |
|---|---|
| Dropout | Neural networks |
| Early Stopping | Neural networks, XGBoost |
| Pruning | Decision trees |
| max_depth | Tree models |
| min_samples_leaf | Tree models |
