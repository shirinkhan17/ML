# 4. Neural Networks & Deep Learning

## What is a Neural Network?

A Neural Network is a machine learning model inspired by the human brain.

It consists of layers of interconnected nodes (neurons) that learn patterns from data.

Structure:

```text
Input Layer
      ↓
Hidden Layer(s)
      ↓
Output Layer
```

Neural networks can learn highly complex and nonlinear relationships.

Applications:

- Computer Vision
- NLP
- Recommendation Systems
- Time Series Forecasting

---

# Neural Network Fundamentals

## Perceptron

The Perceptron is the simplest neural network.

It takes inputs, applies weights, adds a bias, and produces an output.

$$
z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

Output:

$$
\hat y = f(z)
$$

where \(f\) is an activation function.

Limitations:

- Can only learn linearly separable data.

---

## Feed Forward Neural Network (FNN)

A Feed Forward Neural Network contains:

- Input Layer
- One or more Hidden Layers
- Output Layer

Information flows in one direction:

```text
Input → Hidden Layer(s) → Output
```

No loops or cycles exist.

Also called:

- Multi-Layer Perceptron (MLP)

---

## Activation Functions

Activation functions introduce nonlinearity.

Without activation functions, deep networks behave like linear models.

### Sigmoid

$$
\sigma(x)=\frac{1}{1+e^{-x}}
$$

Range:

```text
0 to 1
```

Used for:

- Binary Classification Output Layer

Cons:

- Vanishing gradients

---

### Tanh

$$
tanh(x)
=
\frac{e^x-e^{-x}}
{e^x+e^{-x}}
$$

Range:

```text
-1 to 1
```

Pros:

- Zero-centered

Cons:

- Vanishing gradients

---

### ReLU

$$
ReLU(x)=max(0,x)
$$

Most commonly used activation.

Pros:

- Fast
- Simple
- Reduces vanishing gradients

Cons:

- Dead neurons

---

### Leaky ReLU

$$
f(x)=
\begin{cases}
x, & x>0 \\
0.01x, & x\le0
\end{cases}
$$

Fixes dead neuron problem.

---

### GELU

Used in modern Transformers.

Examples:

- BERT
- GPT

Provides smoother activation than ReLU.

---

## Backpropagation

Backpropagation is the learning algorithm used to train neural networks.

Steps:

1. Forward Pass
2. Compute Loss
3. Compute Gradients
4. Update Weights

Flow:

```text
Forward Pass
      ↓
Compute Loss
      ↓
Backpropagation
      ↓
Update Weights
```

Uses the chain rule from calculus.

Goal:

Minimize the loss function.

---

# Optimization

Optimization means finding weights that minimize loss.

---

## Gradient Descent

Updates weights using gradients.

Update rule:

$$
w_{new}
=
w_{old}
-
\eta
\frac{\partial J}{\partial w}
$$

where:

- \(J\) = Loss Function
- \(\eta\) = Learning Rate

---

## Optimizers

Optimizers improve Gradient Descent.

### SGD (Stochastic Gradient Descent)

Updates weights using one sample at a time.

Pros:

- Simple
- Low memory

Cons:

- Noisy updates

---
## Momentum

Standard Gradient Descent updates weights using only the current gradient.

Problem:

```text
Steep valleys
     ↓
Oscillations
     ↓
Slow convergence
```

Momentum keeps track of previous updates and adds a fraction of them to the current update.

Intuition:

Think of rolling a ball downhill.

```text
Gradient Descent:
Small step
Small step
Small step

Momentum:
Builds speed
Moves faster
```

Benefits:

- Faster convergence
- Reduces oscillations
- Escapes shallow local minima
- Works well for deep networks

Common value:

$$
\beta = 0.9
$$

Momentum accelerates Gradient Descent by maintaining a moving average of past gradients. Instead of updating weights using only the current gradient, it accumulates velocity from previous updates, allowing the optimizer to move faster in consistent directions and reducing oscillations in steep valleys. This leads to faster and more stable convergence.

---

## RMSProp

Different parameters may require different learning rates.

Problem:

```text
One parameter needs large updates
Another parameter needs small updates
```

RMSProp adapts the learning rate for each parameter individually.

Idea:

- Large historical gradients → smaller updates
- Small historical gradients → larger updates

Benefits:

- Handles sparse gradients
- More stable training
- Faster convergence than SGD
- Works well for RNNs

Typical learning rate:

$$
0.001
$$

RMSProp maintains an exponentially weighted moving average of squared gradients and uses it to scale parameter updates. Parameters with large historical gradients receive smaller updates, while parameters with small gradients receive larger updates. This creates adaptive learning rates for each parameter, resulting in faster and more stable training, especially for deep neural networks and recurrent networks.

---

## Adam (Adaptive Moment Estimation)

Adam combines:

1. Momentum (first moment)
2. RMSProp (second moment)

Therefore Adam gets benefits from both:

```text
Momentum
    +
Adaptive Learning Rates
    =
Adam
```

Benefits:

- Fast convergence
- Works well with minimal tuning
- Handles noisy gradients
- Suitable for large datasets
- Default optimizer for most deep learning tasks

Typical parameters:

```python
learning_rate = 0.001
beta1 = 0.9
beta2 = 0.999
```

Used in:

- CNNs
- NLP Models
- Transformers
- Recommendation Systems

For most projects:

```text
Start with Adam
```

is a good rule of thumb.

---

## AdamW

One limitation of Adam is that L2 regularization is not implemented correctly as true weight decay.

AdamW fixes this issue by separating:

```text
Gradient Update
        +
Weight Decay
```

instead of mixing them together.

Benefits:

- Better regularization
- Better generalization
- More stable training
- Often better than Adam on large models

Widely used in:

- BERT
- GPT
- LLaMA
- Modern Transformers

Typical choice:

```python
optimizer = AdamW()
```

for transformer-based models.

---

## Optimizer Comparison

| Optimizer | Main Idea | Advantages | Disadvantages |
|------------|------------|------------|------------|
| SGD | Current gradient only | Simple, low memory | Slow convergence |
| Momentum | Uses previous gradients | Faster, less oscillation | Extra hyperparameter |
| RMSProp | Adaptive learning rates | Stable training | Can overfit |
| Adam | Momentum + RMSProp | Fast, robust, default choice | Sometimes poorer generalization |
| AdamW | Adam + Proper Weight Decay | Better generalization | Slightly more computation |

---

### Summary

- **Momentum** accelerates SGD by using past gradients.
- **RMSProp** adapts learning rates for individual parameters.
- **Adam** combines Momentum and RMSProp and is the most commonly used optimizer.
- **AdamW** improves Adam by implementing proper weight decay and is the standard optimizer for modern Transformer models.

## Learning Rate Scheduling

Learning rate controls update size.

### Large Learning Rate

```text
Fast but unstable
```

### Small Learning Rate

```text
Stable but slow
```

### Scheduling

Gradually decrease learning rate during training.

Examples:

- Step Decay
- Exponential Decay
- Cosine Annealing

Benefits:

- Faster convergence
- Better final accuracy

---

## Weight Initialization

Proper initialization improves training stability.

### Xavier Initialization

Used for:

- Sigmoid
- Tanh

Keeps variance stable across layers.

---

### He Initialization

Used for:

- ReLU
- Leaky ReLU

Most common initialization in modern neural networks.

---

# Regularization

Regularization helps prevent overfitting.

---

## L1 Regularization

Adds penalty:

$$
Loss + \lambda \sum |w_i|
$$

Effects:

- Sparse weights
- Feature selection

---

## L2 Regularization

Adds penalty:

$$
Loss + \lambda \sum w_i^2
$$

Effects:

- Shrinks weights
- More stable models

---

## Dropout

Randomly removes neurons during training.

Example:

```text
Training:
Neuron A → Removed
Neuron B → Active
Neuron C → Active
```

Benefits:

- Prevents co-adaptation
- Reduces overfitting

Typical values:

```text
0.2 – 0.5
```

---

## Batch Normalization

Normalizes activations within a mini-batch.

Benefits:

- Faster training
- More stable gradients
- Allows higher learning rates

Often placed:

```text
Linear Layer
      ↓
BatchNorm
      ↓
Activation
```

---

## Early Stopping

Monitor validation loss.

Stop training when validation performance stops improving.

Example:

```text
Epoch 1 → Better
Epoch 2 → Better
Epoch 3 → Better
Epoch 4 → Worse
Epoch 5 → Worse

Stop
```

Benefits:

- Prevents overfitting
- Saves training time
