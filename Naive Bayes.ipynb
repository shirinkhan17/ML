## Naive Bayes

Naive Bayes is a probabilistic classification algorithm based on Bayes' Theorem.

It predicts the probability that a data point belongs to a particular class and chooses the class with the highest probability.

### Bayes' Theorem


::contentReference[oaicite:0]{index=0}


where:

- \(P(A|B)\) = Posterior Probability
- \(P(B|A)\) = Likelihood
- \(P(A)\) = Prior Probability
- \(P(B)\) = Evidence

---

### Intuition

Suppose we want to classify an email as:

```text
Spam
or
Not Spam
```

Given the words:

```text
"Free"
"Money"
"Offer"
```

Naive Bayes computes:

$$
P(Spam|Words)
$$

and

$$
P(NotSpam|Words)
$$

The class with the larger probability is selected.

---

### Why is it called "Naive"?

The algorithm assumes that all features are conditionally independent given the class.

Example:

For predicting spam,

Naive Bayes assumes:

```text
Word = "Free"
Word = "Money"
Word = "Offer"
```

are independent once we know the email is spam.

Mathematically:

$$
P(x_1,x_2,\ldots,x_n|y)
=
\prod_{i=1}^{n}
P(x_i|y)
$$

This assumption is usually unrealistic because many features are correlated.

However, surprisingly, Naive Bayes often works very well in practice.

---

### Training Process

For each class:

1. Compute class probability

$$
P(y)
$$

2. Compute feature probabilities

$$
P(x_i|y)
$$

3. Apply Bayes' Theorem

4. Select class with highest posterior probability

$$
\hat y
=
argmax_y
P(y|x)
$$

---

### Example

Suppose:

| Email Type | Count |
|------------|--------|
| Spam | 40 |
| Not Spam | 60 |

Then:

$$
P(Spam)=0.4
$$

$$
P(NotSpam)=0.6
$$

If the word "Free" appears:

$$
P(Free|Spam)=0.8
$$

$$
P(Free|NotSpam)=0.1
$$

A new email containing "Free" is much more likely to be classified as spam.

## Variants of Naive Bayes

### 1. Gaussian Naive Bayes

Used when features are continuous.

Assumes features follow a Gaussian (Normal) distribution.

Examples:

- Age
- Salary
- Height
- Weight

Probability density:

$$
P(x|y)
=
\frac{1}{\sqrt{2\pi\sigma^2}}
e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

Applications:

- Medical diagnosis
- Sensor measurements
- Numerical datasets

---

### 2. Multinomial Naive Bayes

Used for count-based features.

Examples:

```text
Word Counts
Document Classification
Email Classification
Sentiment Analysis
```

Feature values:

```text
"free" appears 5 times
"money" appears 3 times
```

Applications:

- Spam detection
- Text classification
- News categorization

Most common Naive Bayes variant in NLP.

---

### 3. Bernoulli Naive Bayes

Used for binary features.

Feature values:

```text
0 = word absent
1 = word present
```

Example:

| Word | Value |
|--------|--------|
| Free | 1 |
| Offer | 1 |
| Meeting | 0 |

Applications:

- Binary text features
- Presence/absence data

---

## Advantages

### Extremely Fast

Training complexity is very low.

Works well on large datasets.

---

### Requires Little Training Data

Can perform well even with limited data.

---

### Handles High-Dimensional Data

Suitable for:

- Text Classification
- Email Filtering
- Document Categorization

where thousands of features may exist.

---

### Probabilistic Output

Provides class probabilities.

Example:

```text
Spam = 0.95
Not Spam = 0.05
```

---

## Disadvantages

### Independence Assumption

The biggest limitation.

Example:

```text
"New"
"York"
```

These words are clearly related.

Naive Bayes assumes they are independent.

This assumption is often incorrect.

---

### Cannot Learn Feature Interactions

Example:

```text
Income + Age
```

may together predict loan approval.

Naive Bayes treats them independently.

---

### Probability Estimates May Be Poor

Class predictions are often good.

Actual probabilities may be poorly calibrated.

---

## When to Use Naive Bayes

Good choice when:

- Dataset is small
- Fast training is needed
- Text classification problems
- Spam detection
- Sentiment analysis
- Document categorization

Avoid when:

- Strong feature interactions exist
- High prediction accuracy is critical
- Features are highly dependent

---

## Interview Summary

Naive Bayes is a probabilistic classifier based on Bayes' theorem that assumes features are conditionally independent given the class label. Despite this strong assumption, it performs surprisingly well for text classification tasks such as spam filtering, sentiment analysis, and document categorization. Common variants include Gaussian Naive Bayes for continuous features, Multinomial Naive Bayes for count data, and Bernoulli Naive Bayes for binary features.
