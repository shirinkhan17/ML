# Natural Language Processing (NLP)

## What is NLP?

Applications:

- Search Engines
- Chatbots
- Machine Translation
- Sentiment Analysis
- Question Answering
- Large Language Models (LLMs)

```text
Input:
"I love machine learning."

Output:
Positive Sentiment
```

# Text Processing

Text data must be converted into a numerical representation before machine learning models can use it.

Typical NLP Pipeline:

```text
Raw Text
    ↓
Tokenization
    ↓
Cleaning
    ↓
Feature Extraction
    ↓
Model Training
```

---

## Tokenization

Tokenization splits text into smaller units called tokens.

Example:

```text
Sentence:
"I love machine learning"
```

Tokens:

```python
["I", "love", "machine", "learning"]
```

Types:

### Word Tokenization

```text
"Machine Learning is fun"

↓

["Machine", "Learning", "is", "fun"]
```

### Sentence Tokenization

```text
"I love NLP. It is interesting."

↓

["I love NLP.", "It is interesting."]
```

### Subword Tokenization

Used in modern LLMs.

Example:

```text
unhappiness

↓

["un", "happi", "ness"]
```

Benefits:

- Handles unseen words
- Smaller vocabulary size

Used in:

- BERT
- GPT
- LLaMA

## Stemming

Stemming reduces words to their root form by removing prefixes and suffixes.

Example:

| Word | Stem |
|--------|--------|
| Playing | Play |
| Played | Play |
| Plays | Play |
| Studies | Studi |

Notice:

```text
Studies → Studi
```

The resulting word may not be a valid English word.

Popular Stemmer:

- Porter Stemmer

Advantages:

- Fast
- Simple

Disadvantages:

- Can produce incorrect words

---

## Lemmatization

Lemmatization converts words to their dictionary (lemma) form.

Example:

| Word | Lemma |
|--------|--------|
| Playing | Play |
| Played | Play |
| Better | Good |
| Studies | Study |

Unlike stemming:

```text
Study
```

remains a valid word.

Advantages:

- More accurate
- Preserves meaning

Disadvantages:

- Slower than stemming

---

## Stemming vs Lemmatization

| Aspect | Stemming | Lemmatization |
|----------|----------|----------|
| Rule Based | Yes | No |
| Uses Vocabulary | No | Yes |
| Output Always Valid Word | No | Yes |
| Speed | Faster | Slower |
| Accuracy | Lower | Higher |

Example:

```text
Studies

Stemming      → Studi
Lemmatization → Study
```

## TF-IDF

TF-IDF stands for:

```text
Term Frequency - Inverse Document Frequency
```

It measures how important a word is in a document relative to a collection of documents.

---

### Term Frequency (TF)

Measures how often a word appears in a document.

$$
TF =
\frac{\text{Word Count}}
{\text{Total Words}}
$$

Example:

```text
Document:
"I love NLP. NLP is fun."

NLP count = 2
Total words = 5
```

$$
TF = \frac{2}{5}
$$

---

### Inverse Document Frequency (IDF)

Measures how rare a word is across documents.

$$
IDF
=
\log
\left(
\frac{N}
{df}
\right)
$$

where:

- \(N\) = total documents
- \(df\) = documents containing the word

Common words:

```text
the
is
and
```

receive low scores.

Rare words receive high scores.

---

### TF-IDF Score

$$
TFIDF = TF \times IDF
$$

Interpretation:

```text
High TF
+
Rare Word
=
High TF-IDF
```

Applications:

- Search Engines
- Text Classification
- Document Ranking

Limitations:

- No semantic meaning
- Ignores word order

# Word Embeddings

Traditional approaches:

```text
TF-IDF
One-Hot Encoding
```

treat words independently.

Word embeddings learn dense vector representations where similar words have similar vectors.

Example:

```text
King  → [0.2, -0.4, 0.8, ...]
Queen → [0.3, -0.5, 0.7, ...]
```

Words with similar meanings appear close together in vector space.

---

## Word2Vec

Introduced by Google in 2013.

Goal:

Learn word representations based on surrounding context.

Two architectures:

### CBOW (Continuous Bag of Words)

Predict target word from surrounding words.

Example:

```text
I love _____ learning

↓

machine
```

---

### Skip-Gram

Predict surrounding words from target word.

Example:

```text
Target:
machine

Predict:
love
learning
```

Advantages:

- Efficient
- Captures semantic relationships

Famous example:

```text
King - Man + Woman ≈ Queen
```

Limitations:

- One vector per word
- Cannot handle multiple meanings

Example:

```text
bank
```

river bank vs financial bank

same embedding

---

## GloVe

Global Vectors for Word Representation.

Developed by Stanford.

Unlike Word2Vec:

```text
Word2Vec → Local Context

GloVe → Global Word Statistics
```

Uses word co-occurrence matrices.

Example:

```text
King and Queen
```

appear in similar contexts and receive similar embeddings.

Advantages:

- Captures semantic relationships
- Uses global corpus information

Limitations:

- Static embeddings
- One vector per word

---

## FastText

Developed by Meta.

Improvement over Word2Vec.

Instead of learning embeddings for entire words:

```text
Word
↓
Character N-Grams
```

Example:

```text
learning

↓

lea
ear
arn
rni
nin
ing
```

Advantages:

### Handles Rare Words

Can generate embeddings for unseen words.

Example:

```text
machinelearning
```

may never appear during training but can still receive an embedding.

### Handles Misspellings

Example:

```text
colour
color
```

receive similar representations.

### Better for Morphologically Rich Languages

Examples:

- Turkish
- Finnish
- Arabic

## Key Summary

### Text Processing

- Tokenization splits text into tokens.
- Stemming removes suffixes and prefixes.
- Lemmatization converts words to valid dictionary forms.
- TF-IDF measures word importance.

### Word Embeddings

- Word2Vec learns embeddings using local context.
- GloVe learns embeddings using global word statistics.
- FastText learns embeddings using character n-grams.

### Important Limitation

Word2Vec, GloVe, and FastText are **static embeddings**:

```text
bank
```

has one vector regardless of context.

Modern models such as BERT generate **contextual embeddings**, where the meaning depends on surrounding words.
